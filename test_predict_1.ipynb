```python
from ultralytics import YOLO
import cv2
import numpy as np
import os
model = YOLO("result/best.pt") 
link_ = "data/1260/rgb/"
```

```python
for i in os.listdir(link_):
    frame = cv2.imread(link_+i)
    height,width,_ = frame.shape
    results = model(frame)
    box_ = np.array([])
    frame = cv2.resize(frame,(int(width*5),int(height*5)))
    for result in results:
        detections = []
        for result in results:
            for i in range(len(result.boxes.xyxy)):
                x_min, y_min, x_max, y_max = np.array(result.boxes.xyxy)[0]

                score = round(float(result.boxes.conf[0]),2)
                # Tracking ch ly cc bbox c  tin cy cao
                if score > 0.7:
                    detections.append([x_min, y_min, x_max, y_max, score])
                    box_ = result.boxes.xywh[0].numpy()
                    cls_ = result.boxes.cls[0].numpy()
                    box_ = box_ / np.array([width, height, width, height])

        # Convert thnh numpy array  input vo SORT
        detections = np.array(detections)
        
        if(detections.size > 0):

            # V cc bounding box v ID
            for obj in detections:
                x_min, y_min, x_max, y_max, _ = obj.astype(int)*5
                if(result.boxes.cls.numpy()[0]==0):
                    cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
                    cv2.putText(frame, f"Role: {cls_} Score: {score}", (x_min, y_min - 10),
                                cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 255, 0), 2)
                else:
                    cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 0, 255), 2)
                    cv2.putText(frame, f"Role: {cls_} Score: {score}", (x_min, y_min - 10),
                                cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 0, 255), 2)
        cv2.imshow("Tracking", frame)
        cv2.waitKey(1)
cv2.destroyAllWindows()
```
